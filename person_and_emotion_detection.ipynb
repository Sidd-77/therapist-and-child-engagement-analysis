```python
import cv2
from deepface import DeepFace
```

```python
# Initialize dictionary to store emotion counts
emotion_counts = {}
cap = cv2.VideoCapture('videos/3.mp4')
frame_count = 0
from mtcnn import MTCNN

# Initialize the MTCNN model
detector = MTCNN()


while(cap.isOpened()):
    ret, frame = cap.read()

    if ret:
        frame_count += 1

        # Process every 5th frame
        if frame_count % 5 == 0:
            # Detect faces
            faces = detector.detect_faces(frame)

            for i, face in enumerate(faces):
                # Extract bounding box and keypoints
                x, y, w, h = face['box']
                keypoints = face['keypoints']

                # Extract face
                face_img = frame[y:y+h, x:x+w]

                # Analyze emotions
                results = DeepFace.analyze(face_img, actions=['emotion'], enforce_detection=False)

                for result in results:
                    dominant_emotion = result['dominant_emotion']
                    # rest of your code

                # Update emotion count
                if dominant_emotion in emotion_counts:
                    emotion_counts[dominant_emotion] += 1
                else:
                    emotion_counts[dominant_emotion] = 1

                # Display dominant emotion
                cv2.putText(frame, dominant_emotion, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

                # Draw rectangle around the face
                cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

            # Display frame
            cv2.imshow('Video', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()

# Print emotion counts
for emotion, count in emotion_counts.items():
    print(f'{emotion}: {count}')
```
